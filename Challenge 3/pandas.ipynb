```python
import pandas as pd
```

# Part a

```python
df = pd.DataFrame({'days': [1, 1, 2, 2, 1, 3, 4],
                   'values': [10, 10, 5, 3, -2, 4, 20]})
df
```

```python
df1 = df.groupby('days') \
    .aggregate(['mean', 'median', 'max', 'min']) \
    .reset_index()
df1.columns = ['days', 'mean_values', 'median_values', 'max_values', 'min_values']
df1
```

# Part b

```python
df = pd.DataFrame({'employee': [1001, 1002, 1004, 1001, 1001, 1002, 1004, 1005, 1005],
                       'pos': [2, 2, 2, 2, 2, 2, 2, 2, 2],
                       'amount': [125, 542, 2345, 892, 100, 1234, 657, 34, 35]})
df
```

```python
df2 = df.groupby('employee') \
    .aggregate({'pos': ['mean'], 'amount':['max', 'min']}) \
    .reset_index()
df2.columns = ['employee', 'pos', 'max', 'min']
df2['amount_diff'] = df2['max'] - df2['min']
df2 = df2.drop(columns=['max', 'min'])
df2.sort_values('amount_diff', ascending=False)[:2].reset_index()
```
